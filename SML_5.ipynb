```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

data=pd.read_csv(r"D:\ISI 2025\Class\Subjects\Machine Learning\data.csv")

data
```

```python
data_X=data.iloc[:,1:10]
data_X
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

WCSS_values=[]
k_values=[]

for k in range(1,25):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_X)
    WCSS=round(kmeans.inertia_,4)
    k_values.append(k)
    WCSS_values.append(WCSS)

# Plot k values vs. WCSS
plt.figure(figsize=(6,4))
plt.plot(k_values,WCSS_values,)
plt.title("Elbow method for optimal k")
plt.xlabel("Number of clusters(k)")
plt.ylabel("WCSS")
plt.show()
```

```python
from pyclustering.cluster.kmedoids import kmedoids
import numpy as np

def SS_medoid(k):
    
    # Convert DataFrame to NumPy array
    data_X_array = data_X.values
    
    # Randomly initialize medoids
    initial_medoids=np.random.choice(len(data_X),k,replace=False).tolist()

    # Perform PAM clustering
    PAM = kmedoids(data_X_array, initial_medoids)
    PAM.process()

    # Get medoids
    medoids = PAM.get_medoids()

    # Get Clusters
    clusters = PAM.get_clusters()

    # Extract the medoid points
    medoid_points = data_X_array[medoids]

    L_SS=[]
    
    # Extract the clusters and sum of squares
    for i in range(len(clusters)):
        Sum_of_squares_Cluster= np.sum((data_X_array[clusters[i]]-medoid_points[i])**2)
        L_SS.append(Sum_of_squares_Cluster)
    SS=0
    for j in range(len(L_SS)):
        SS=SS+L_SS[j]
    return SS
```

```python
## Sum of Suares for each clusters
SS_for_diff_K=[]
K_values=[]

for k in range(1,26):
    SS_for_diff_K.append(SS_medoid(k))
    K_values.append(k)

# Plot k values vs. SS for k medoid:
plt.figure(figsize=(6,4))
plt.plot(K_values,SS_for_diff_K)
plt.title("Elbow method for optimal k")
plt.xlabel("Number of clusters(k)")
plt.ylabel("SS")
plt.show()
```

### Normalize the data between [0,1]

```python
data_range=pd.DataFrame()

for i in data_X.columns:
    range_N = []
    #print(list(data_X[i]))
    # Calculate min and max values of 'person_age'
    minimum = min(data_X[i])
    maximum = max(data_X[i])

    # Loop through the column and normalize each value
    for j in data_X[i]:
        r = round((j - minimum) / (maximum - minimum), 4)  # Normalize using range
        range_N.append(r)
    data_range[i]=range_N
    #print(range)
data_range 
```

```python
#Strip whitespace from the column names:

data_range.columns = data_X.columns.str.strip()
data_range["Eggs"]
```

```python
## Elbow for Means for normalized data

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

Wcss_values_=[]
k_values_=[]

for k in range(1,25):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_range)
    Wcss_values_range=round(kmeans.inertia_,4)
    k_values_.append(k)
    Wcss_values_.append(Wcss_values_range)

# Plot k values vs. WCSS
plt.figure(figsize=(6,4))
plt.plot(k_values_,Wcss_values_)
plt.title("Elbow method for optimal k")
plt.xlabel("Number of clusters(k)")
plt.ylabel("WCSS")
plt.show()
```

```python
# Define the KMeans model with 10 clusters and 5 initializations (nstart=5 equivalent)
kmeans = KMeans(n_clusters=5, n_init=5)

# Fit the model
kmeans.fit(data_range)

# Add the cluster labels to the original dataset
data_range['cluster'] = kmeans.labels_

data_range
```

## Map with clusters

```python
Clusters=[]
for i in range(0,5):
    l=[]

    g = data_range.groupby('cluster')
    index=g.get_group(i).index

    # Extract values from a particular column, e.g., 'column_name', using the indexes
    countries = data.loc[index,'Country'] #Filters the rows based on the index and retrieves the Country column.
    countries.tolist() #Converts the pandas Series (with index) into a plain Python list containing only the values.

    l.extend(countries) # Adds the country names directly to the list l
    Clusters.append(l)
Clusters
```
